```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline

def gradient_descent(x,y) :
    m=b=0
    i=10000
    n=len(x)
    lr=0.0002
    
    for j in range(i):
        yp=m*x+b
        cost=(1/n)*sum([val**2 for val in (y-yp)])
        md=-(2/n)*sum(x*(y-yp))
        bd=-(2/n)*sum(y-yp)
        m=m-lr*md
        b=b-lr*bd
        
    
    return m,b


x = np.array(range(0,9))
y = np.array([35,40,25,30,31,22,24,37,38])

m,b = gradient_descent(x,y)
```
